## Data cleaning

### Subtask:
Clean the Louis Vuitton stock price data by handling any outliers and addressing gaps in the date sequence.


**Reasoning**:
Handle outliers in the price columns ('Open', 'High', 'Low', 'Close', 'Adj Close') using winsorization, and address gaps in the date sequence using forward fill. Then validate the cleaned data.



```python
import pandas as pd
import numpy as np

# Outlier Handling (Winsorization)
price_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close']
for col in price_cols:
    df[col] = np.clip(df[col], df[col].quantile(0.01), df[col].quantile(0.99))

# Date Sequence Gaps Handling (Forward Fill)
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df = df.resample('D').ffill()
df = df.reset_index()

# Data Validation
print(df.describe())
print(df.isnull().sum())
print(df['Date'].diff().value_counts())

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Louis Vuitton Close Price Over Time (After Cleaning)')
plt.grid(True)
plt.show()
```

## Data wrangling

### Subtask:
Engineer relevant features from the cleaned Louis Vuitton stock data.


**Reasoning**:
Calculate the required features (moving averages, RSI, volatility, daily return) and handle potential NaN values using forward fill.



```python
import pandas as pd
import numpy as np

# Calculate Moving Averages
df['MA7'] = df['Close'].rolling(window=7).mean()
df['MA21'] = df['Close'].rolling(window=21).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()

# Calculate RSI
delta = df['Close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
df['RSI14'] = 100 - (100 / (1 + rs))
df['RSI14'] = df['RSI14'].fillna(0) # Fill initial NaN with 0
df['RSI14'] = np.clip(df['RSI14'], 0, 100) # Clip values to 0-100 range


# Calculate Volatility
df['Volatility21'] = df['Close'].rolling(window=21).std()

# Calculate Percentage Change
df['Daily_Return'] = df['Close'].pct_change()

# Forward fill NaN values for all new columns
for col in ['MA7', 'MA21', 'MA50', 'Volatility21', 'Daily_Return']:
    df[col] = df[col].ffill()

# Display first few rows to check the results
display(df.head(25))
```

## Data analysis

### Subtask:
Perform time series analysis on the Louis Vuitton stock data to identify trends, seasonality, and potential patterns.


**Reasoning**:
Analyze the time series, detect seasonality, explore autocorrelations, and summarize findings with visualizations.



```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 1. Trend Analysis
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['Close'].rolling(window=30).mean(), label='30-Day Rolling Mean')
plt.plot(df['Date'], df['Close'].rolling(window=30).std(), label='30-Day Rolling Std')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Louis Vuitton Close Price with Rolling Statistics')
plt.legend()
plt.grid(True)
plt.show()

# Decomposition
decomposition = seasonal_decompose(df['Close'], model='additive', period=365) # Assuming yearly seasonality
fig = decomposition.plot()
fig.set_size_inches(14, 7)
plt.show()

# 2. Seasonality
# The decomposition plot above shows seasonality.

# 3. Autocorrelation Analysis
plt.figure(figsize=(12, 6))
plot_acf(df['Close'], lags=50, ax=plt.gca()) # Autocorrelation function
plt.title('ACF of Louis Vuitton Close Price')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(df['Close'], lags=50, ax=plt.gca()) # Partial Autocorrelation function
plt.title('PACF of Louis Vuitton Close Price')
plt.show()

# 4. Summary
print("Summary of Time Series Analysis:")
print("- Trend: The rolling statistics indicate some long-term trends but also some volatility.")
print("- Seasonality: The decomposition plot shows a clear yearly seasonality pattern.")
print("- Autocorrelation: The ACF and PACF plots help determine significant lags for potential ARIMA modeling.")
```
